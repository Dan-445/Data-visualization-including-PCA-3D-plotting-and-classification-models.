````python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()

# Displaying the data structure
print(breast_cancer.keys())
print("Data shape:", breast_cancer.data.shape)
print("Target shape:", breast_cancer.target.shape)
print("Feature names:", breast_cancer.feature_names)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.25, random_state=0)

# Creating and training the KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predicting and evaluating the model
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))

# Confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(confusion_matrix(y_test, y_pred), interpolation='nearest', cmap=plt.cm.Greens)
plt.title('Confusion matrix for Breast Cancer dataset')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
````

````python
from sklearn.decomposition import PCA

# PCA to reduce dimensions for visualization
pca = PCA(n_components=2)
X_r = pca.fit_transform(breast_cancer.data)

# Plotting
plt.figure()
colors = ['darkorange', 'navy']
target_names = ['Malignant', 'Benign']
lw = 2

for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(X_r[breast_cancer.target == i, 0], X_r[breast_cancer.target == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Breast Cancer dataset')
plt.show()
````

````python
from sklearn.datasets import load_wine
wine = load_wine()

# Displaying the data structure
print(wine.keys())
print(wine.data.shape)
print(wine.target.shape)
````

````python
# Splitting data
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=0)

# Creating and training the KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predicting and evaluating the model
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))

# Confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(confusion_matrix(y_test, y_pred), interpolation='nearest', cmap=plt.cm.Purples)
plt.title('Confusion matrix for Wine dataset')
plt.colorbar()
plt.show()
````

````python
from sklearn.decomposition import PCA

# PCA to reduce dimensions for visualization
pca = PCA(n_components=2)
X_r = pca.fit_transform(wine.data)

# Plotting
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], wine.target_names):
    plt.scatter(X_r[wine.target == i, 0], X_r[wine.target == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Wine dataset')
plt.show()
````
